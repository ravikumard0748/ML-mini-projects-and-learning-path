```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
cancer.target
```

```python
data=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
data.info()
```

```python
tar=pd.DataFrame(cancer['target'],columns=['Cancer'])
tar.head()
```

```python
data.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
nd=scale.fit_transform(data)
nd
```

```python
nd
```

```python
nd=pd.DataFrame(nd,columns=cancer['feature_names'])
```

```python
nd.head()
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(nd,tar,test_size=0.30,random_state=105)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=1)
```

```python
knn.fit(xtrain,ytrain)
```

```python
pre=knn.predict(xtest)
```

```python
from sklearn.metrics import *
```

```python
clas=classification_report(ytest,pre)
conf=confusion_matrix(ytest,pre)
```

```python
print(conf,"\n",clas)
```

```python
eror_rate=[]
ytrain = ytrain.reshape(-1,1)
ytest = ytest.ravel(-1,1)
for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(xtrain,ytrain)
    pred_i=knn.predict(xtest)
    eror_rate.append(np.mean(pred_i!=ytest))
print(eror_rate)
```

```python
print(len(range(1,41)))
print(len(eror_rate))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,41),eror_rate,color='blue',linestyle='dashed',marker='o',maekerfacecolor='red',markersize=10)
plt.title('Error Rate vs K value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```
