```python
from sklearn.cluster import KMeans
```

```python
model=KMeans(n_clusters=5)
```

```python
newdata.isnull().sum()
```

```python
newdata.dropna(inplace=True)
```

```python
newdata.isnull().sum()
```

```python
 model.fit(newdata)
```

```python
print(model.labels_)
```

```python
print(len(model.labels_))
```

```python
unique,counts=np.unique(model.labels_,return_counts=True)
print(dict(zip(unique,counts)))
```

```python
newdata['clusters']=model.labels_
sns.set_style('whitegrid')
sns.lmplot(x='director_facebook_likes', 
           y='actor_3_facebook_likes', 
           data=newdata, 
           hue='clusters', 
           palette='coolwarm', 
           height=6, 
           aspect=1, 
           fit_reg=False)
```
