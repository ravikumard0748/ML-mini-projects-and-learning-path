```python
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset
diabetes = load_diabetes()
data = diabetes.data  # Feature matrix (NumPy array)
target = diabetes.target  # Target vector (NumPy array)

# Split the data into training and testing sets
# We use train_test_split for a more standard approach
data_train_x = data[:-30]  # All rows except the last 30 for training features
data_test_x = data[-20:]   # Last 20 rows for testing features

data_train_y = target[:-30]  # All rows except the last 30 for training targets
data_test_y = target[-20:]   # Last 20 rows for testing targets

# Optionally, you can use train_test_split to create random splits
# data_train_x, data_test_x, data_train_y, data_test_y = train_test_split(data, target, test_size=0.2, random_state=42)

# Feature scaling (optional but recommended)
scaler = StandardScaler()
data_train_x = scaler.fit_transform(data_train_x)
data_test_x = scaler.transform(data_test_x)

# Train a regression model
model = LinearRegression()
model.fit(data_train_x, data_train_y)

# Make predictions
predictions = model.predict(data_test_x)

# Evaluate the model
mse = mean_squared_error(data_test_y, predictions)
r2 = r2_score(data_test_y, predictions)

print(f"Mean Squared Error: {mse}")
print(f"R^2 Score: {r2}")
```
